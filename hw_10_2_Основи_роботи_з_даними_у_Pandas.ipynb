7.2 Сформулюйте 3 гіпотези, чому в одному місті gross income вище, як можна дізнатись чи правдива ця гіпотеза з використанням даних (наявних і додаткових), та як висновки з цих гіпотез можна використати для прийняття рішень як підвищити прибуток в мережі супермаркетів. Відповідь напишіть текстом в окремій клітинці типу Markdown.

Це завдання - на продуктове мислення. Якщо не знаєте, які гіпотези сформулювати - напишіть ті, які здаються вам зовсім примітивними. Важливо навчитись висловлювати свої гіпотези щодо даних, і з часом та досвідом вони будуть ставати щоразу глибше і точніше. Весь аналіз даних будується на тому, що ми перевіряємо певні гіпотези та робимо висновки і рекоемндації на основі своїх досліджень.

```python
df.rename(columns={'gross income' : 'gross_income'}, inplace=True)
```

```python
filter_1 = df.City =='Yangon'
yangon_count = df[filter_1].gross_income.count()
filter_1 = (df.City =='Yangon')
yangon_sum = df[filter_1].gross_income.sum()
yangon_avg = yangon_sum / yangon_count
print(f'Yangon gross income = {yangon_sum}\nYangon count = {yangon_count} \nYangon gross income AVG = {yangon_avg}')
```

```python
filter_2 = df.City =='Mandalay'
mandalay_count = df[filter_2].gross_income.count()
filter_2 = (df.City =='Mandalay')
mandalay_sum = df[filter_2].gross_income.sum()
mandalay_avg = mandalay_sum / mandalay_count
print(f'Mandalay gross income = {mandalay_sum}\nMandalay count = {mandalay_count} \nMandalay gross income AVG = {mandalay_avg}')
```

В Yangon незначно більшим є сукупний gross income, а в Mandalay більшим є gross income AVG. 
Можливі причини більшого загального gross income (в Yangon):
1) більше покупців (або покупок)
2) продано більше товарів на покупця (більшим є середній чек)
3) більше торгових годин в Yangon.
Причини більшого середнього gross income (в Mandalay):
1) в Mandalay продається більше дорожчих категорій товарів
2) в Mandalay менша кількість покупців (покупок) при однаковому рівні виручки
3) в Mandalay gross income AVG = 15.23 більш зручне розташування торгівельних точок, відсутні конкуренти, тому покупці витрачають більше саме в магазинах цієї мережі.
